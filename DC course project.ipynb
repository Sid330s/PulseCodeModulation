```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Here we are Initializing the Frequencies and the Sampling counts
fs = 80
f = 5
quantLevel = 8
quantLevelLst = [-8, -6, -4, -2, 2, 4, 6, 8]
samples = 17
x = np.arange(samples)
y = np.sin(2*(np.pi)*f*(x/fs))
binForm = {-8 : [0,0,0], -6 : [0,0,1], -4 : [0,1,0], -2 : [0,1,1], 2 : [1,0,0], 4 : [1,0,1], 6 : [1,1,0], 8 : [1,1,1]}
```

```python
# Ploting the Simple Sin Wave first
plt.plot(x, y)
plt.xlabel('Time (s)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
# Ploting the Sampled Output
plt.stem(x, y, 'r')
plt.plot(x,y)
plt.xlabel('Time (s)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
def findQuant(temp):
  temp *= 10
  if temp<-7:
    return -8
  elif temp<-5:
    return -6
  elif temp<-3:
    return -4
  elif temp<0:
    return -2
  elif temp<3:
    return 2
  elif temp<5:
    return 4
  elif temp<7:
    return 6
  return 8
```

```python
encodedDigitalSignal = []
quantizedSignal = []
for i in y:
  temp = findQuant(i)
  quantizedSignal.append(temp)
  encodedDigitalSignal += binForm[temp]
```

```python
print(quantizedSignal)
plt.stem(x, quantizedSignal, 'r')
plt.xlabel('Time (s)')
plt.ylabel('voltage(V)')
plt.show()
```

```python
plt.plot(encodedDigitalSignal, drawstyle='steps-pre')
plt.title("Encoded Digital Signal")
plt.show()
```
